```python
import numpy as np
import matplotlib.pyplot as plt


def sigmoid(x):
    return 1 / (1 + np.exp(-x))


x = np.random.randn(1000, 100)  # 1000개의 데이터
node_num = 100  # 각 은닉층의 노드 수
hidden_layer_size = 5  # 은닉층이 5개
activations = {}  # 활성화값을 저장

for i in range(hidden_layer_size):
    if i != 0:
        x = activations[i-1]
    w = np.random.randn(node_num, node_num) * 1
    a = np.dot(x, w)
    z = sigmoid(a)
    activations[i] = z

```

```python
# 서브플롯 설정
num_activations = len(activations)
cols = 3  # 한 줄에 표시할 subplot 개수
rows = (num_activations + cols - 1) // cols  # 필요한 행 수 계산

fig, axes = plt.subplots(rows, cols, figsize=(15, 5 * rows))
axes = axes.flatten()  # axes 배열을 평탄화

# 각 키와 값을 순회하며 히스토그램 그리기
for i, (layer, activation) in enumerate(activations.items()):
    axes[i].hist(activation, bins=10, alpha=0.7)
    axes[i].set_title(layer)  # subplot 제목으로 dictionary의 키 사용
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequency')
    axes[i].set_xlim(0, 1)  # x축을 0에서 1로 고정

# 사용하지 않는 subplot 숨기기
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

# 전체 레이아웃 조정
plt.tight_layout()
plt.show()
```
